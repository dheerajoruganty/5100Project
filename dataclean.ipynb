```python
import pandas as pd
```

```python
df = pd.read_csv('who_dataset_v1.csv')
```

```python
df.head()
```

```python
df = df.loc[(df['NO2 temporal coverage (%)']  > 90)]
```

```python
df = df.iloc[:, :-6]
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
columns = ['PM25 temporal coverage (%)', 'PM10 temporal coverage (%)', 'PM2.5 (μg/m3)', 'PM10 (μg/m3)']
df = df.drop(columns=columns)
```

```python
df
```

```python
df2 = pd.read_csv('who_population.csv')
```

```python
df2.info()
```

```python
df2.columns.tolist()
```

```python
df2.head()
```

```python
merged_df = pd.merge(df, df2)
```

```python
merged_df
```

```python
merged_df.to_csv('data_final.csv')
```

```python
df = merged_df
```

```python
average_no2 = df.groupby(['WHO Country Name', 'Measurement Year'])['NO2 (μg/m3)'].mean()
```

```python
average_no2
```

```python
average_no2_df = average_no2.reset_index()
average_no2_df = average_no2_df.rename(columns={'NO2 (μg/m3)': 'Average NO2'})
```

```python
average_no2_df
```

```python
merged_df = pd.merge(df, average_no2_df)
```

```python
merged_df
```

```python
df = merged_df
```
