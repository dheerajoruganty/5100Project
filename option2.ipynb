```python
from scipy import stats
import pandas as pd

sample1 = df[df['name'] == 'Nitrogen dioxide (NO2)']['data_value']
sample2 = df[df['name'] == 'Ozone (O3)']['data_value']

sample1 = pd.to_numeric(sample1.replace(',', '', regex=True), errors='coerce')
sample2 = pd.to_numeric(sample2.replace(',', '', regex=True), errors='coerce')

# Drop NaN values
sample1 = sample1.dropna()
sample2 = sample2.dropna()

# Perform t-test
t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)

# Print the results
print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the means.")




```
